```python
#import plotly.express as px
# Q39 Plot a line chart that shows the number of gold medals won by the United States between 2000 and 2020.
# ANS The below line graph shows the no of gold medals won by the United States between 2000 and 2020.
df9=df1[(df1['team'] == 'United States')& (df1['medal']=='Gold') & (df1['year'] > 2000) & (df1['year']<2020)].groupby(['team', 'year'])[['medal']].count()
x_axis=['2002','2004','2006','2008','2010','2012','2014','2016','2018']
y_axis=df9['medal']
plt.plot(x_axis, y_axis)
plt.title('Gold medals won by the United States between 2000 and 2020')
plt.xlabel('Year')
plt.ylabel('Count of medals')
plt.show()
```

```python
# Q40 Plot a grouped bar chart that shows the total number of medals won by Norway, the United States, and Canada during 2010, 2014, and 2018 winter Olympics. 
# ANS The below bar graph shows the no of  medals won by the United States, Norway, Canada during 2010,2014 & 2018.
rs22 = df1[((df1['team'] == 'Norway') | (df1['team'] == 'United States') | (df1['team'] == 'Canada')) &( (df1['year']==2010) | (df1['year']==2014) | (df1['year']==2018))]
rs23=pd.crosstab([rs22.team], [rs22.year], values = rs22.medal, aggfunc= 'count')
rs23.plot(kind="bar",title="Medals won by each team", width=0.5)
plt.xlabel("team")
plt.ylabel("medals")
plt.show()
```

```python
# Q41 Is there any correlation between the age or gender of the competing athletes and the type of medals won? 
# ANS I categorized each of my variables into numbers by using cat.codes as medal is a string type. The correlation between age and medal is -0.015125 
#     If the age is inversly proportional to medals if the age increases the no of medals decreases. 
df1['birth_year'] = df1['birth_year'].fillna(0).astype(int)
df1['age']=df1['year']-df1['birth_year']
df1['age']=df1['age'].astype('category').cat.codes
#df1['sex']=df1['sex'].astype('category').cat.codes
df1['medal']=df1['medal'].astype('category').cat.codes
df1.corr()
```

```python
# Q42 Modelling the relationship between dependent and independent variables, create a linear regression model to make a prediction. Choose your own variables from the
#      Olympics dataset.
# ANS  Here i have used age and medal as dependent and independent variables to create a linear regression model. coefficients = -95.52963513, Mean squared error: 15448172.17, 
#      Coefficient of determination: 0.33  

data = df1[(df1.medal != 'nan') & (df1.age >= 0)].groupby(['age'],as_index=False)['medal'].count()
data.shape[0]
data.head()
x = data.age.values.reshape(99,1)
y= data.medal.values.reshape(99,1)
regr = LinearRegression().fit(x, y)
plt.scatter(x, y,  color='red')
plt.plot(x, regr.predict(x), color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
y_pred = regr.predict(x)
y_pred
print("Coefficients: \n", regr.coef_)
print("Mean squared error: %.2f" % mean_squared_error(y, y_pred))
print("Coefficient of determination: %.2f" % r2_score(y, y_pred))
plt.scatter(x, y, color="black")
plt.plot(y, y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
